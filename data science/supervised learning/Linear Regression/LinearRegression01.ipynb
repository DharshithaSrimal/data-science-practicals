```python
#Import dependancies
#Library for scientific computing and calculations.
import numpy as np
#library for plotting data
import matplotlib.pyplot as plt
```

```python
# Generate some data
# np.linspace() function takes three arguments
# start: The starting value, stop: The stopping value, num: The number of samples to generate 
x = np.linspace(0, 10, 10)
y = 2*x + 5
```

```python
# Fit a linear regression model
# polyfit() function returns the coefficients of the linear regression model, which are stored in the variables m and b. 
# The coefficient m = slope of the line, and the coefficient b = y-intercept
m, b = np.polyfit(x, y, 1)
```

```python
# Plot the data and the fitted line
plt.plot(x, y, 'o')
plt.plot(x, m*x + b)
plt.show()
```
